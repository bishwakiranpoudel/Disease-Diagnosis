# Splitting Training and testing data

```python
X = df.drop('prognosis' ,axis=1)
y = df['prognosis']
```

```python
y.head(5)
```

### Label Encoding

```python
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
Y
```

Here, 15 means Fungal infection and 27 means impetigo and soo on..

```python
X_test,X_train,y_test,y_train = train_test_split(X,Y,test_size =0.25 , random_state=42)
```

```python
X_test.shape,X_train.shape,y_test.shape,y_train.shape
```

```python
models = {
    'SVC': SVC(kernel='linear'),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5),
    'LogisticRegression': LogisticRegression(random_state=42),
    'DecisionTree': DecisionTreeClassifier(random_state=42)
}
for model_name , model in models.items():
    # Train model
    model.fit(X_train,y_train)
    # test model
    predictions = model.predict(X_test)
    # calculate accuracy
    accuracy = accuracy_score(y_test,predictions)
    # calculate confusion matrix
    cm = confusion_matrix(y_test,predictions)
    # print results
    print(f"{model_name} accuracy : {accuracy}")
    print(f"{model_name} confusion matrix :")
    print(np.array2string(cm,separator=', '))
```

Suprsingly ,On above cell all model gives the best 100% accuracy so we dont need to go with all model. So i go with Random Forest model

```python
rf = RandomForestClassifier(random_state=42,n_estimators=100)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
# Path to save the model
path = './randomforest.pkl'

# Ensure the directory exists
os.makedirs(os.path.dirname(path), exist_ok=True)

# Save the model
with open(path, 'wb') as file:
    pickle.dump(rf, file)
```

```python
# load the data-set
rf = pickle.load(open(path,'rb'))
```
